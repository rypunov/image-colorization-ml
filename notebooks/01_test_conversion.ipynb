```python
import sys
sys.path.append('../src')  # Чтобы импортировать из src/

#noinspection PyUnresolvedReferences
from data_preparation import rgb_to_gray
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Путь к тестовой фотографии
test_image_path = '../data/raw/pencils.jpg'
color_img = Image.open(test_image_path)

print(f"Размер: {color_img.size}")
print(f"Режим: {color_img.mode}")
```

```python
# Применяем все три метода
gray_average = rgb_to_gray(color_img, method='average')
gray_luma = rgb_to_gray(color_img, method='luma')
gray_pil = rgb_to_gray(color_img, method='pil')

# Отображаем результаты
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Оригинал
axes[0, 0].imshow(color_img)
axes[0, 0].set_title("Оригинал (RGB)")
axes[0, 0].axis('off')

# Average
axes[0, 1].imshow(gray_average, cmap='gray')
axes[0, 1].set_title("Average метод")
axes[0, 1].axis('off')

# Luma
axes[1, 0].imshow(gray_luma, cmap='gray')
axes[1, 0].set_title("Luma метод")
axes[1, 0].axis('off')

# PIL
axes[1, 1].imshow(gray_pil, cmap='gray')
axes[1, 1].set_title("PIL convert('L')")
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Сохраняем для отчёта
gray_average.save('../data/processed/pencils_gray_average.jpg')
gray_luma.save('../data/processed/pencils_gray_luma.jpg')
gray_pil.save('../data/processed/pencils_gray_pil.jpg')
print("Результаты сохранены в data/processed/")
```

## Выводы

1. **Average метод** - самый простой, но может терять контраст
2. **Luma метод** - учитывает восприятие глаза, наиболее естественный
3. **PIL convert('L')** - стандартный, быстрый, хорошее качество

Для проекта будем использовать **Luma метод**.
